```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Generate random points.
```

```python
from sklearn.datasets import make_blobs
```

```python
points, cluster_indexes = make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=0)
x = points[:, 0]
y = points[:, 1]
```

```python
plt.scatter(x, y, s=50, alpha=0.7) # s: size of the markers; alpha: transparency of the markers
```

```python
# Use K-means clustering algorithm to find out the centers of clusters.
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(points)
predicted_cluster_indexes = kmeans.predict(points)
```

```python
plt.scatter(x, y, c=predicted_cluster_indexes, s=50, alpha=0.7, cmap='viridis') # c: color; cmap: color map instance
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100)
```
